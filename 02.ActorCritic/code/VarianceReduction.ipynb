```python
import numpy as np
import seaborn as sns
from scipy.stats import norm
sns.set(style="darkgrid")
```

```python
Rs = np.random.randn(1000)*2 + 100
dist = norm(loc = 100, scale = 2)
pis = np.array([dist.pdf(R) for R in Rs])
```

```python
ax = sns.distplot(Rs,fit=norm, kde = False, fit_kws={"color": "k", "lw": 3},
            hist_kws={"linewidth": 4, 
                    "color": "g"})
ax.set_title("Variance: "+ str(np.std(Rs)**2))
fig = ax.get_figure()
fig.savefig('return_dist.png')
```

```python
(Rs * pis).max()
```

```python
ax = sns.distplot(Rs * (1 - pis),fit=norm, kde = False, fit_kws={"color": "k", "lw": 3},
            hist_kws={"linewidth": 4, 
                    "color": "g"})
ax.set_title("Variance: "+ str(np.std(Rs*(1-pis))**2))
fig = ax.get_figure()
fig.savefig('score_dist.png')
```

```python
Rs_norm = Rs - 100
```

```python
ax = sns.distplot(Rs_norm,fit=norm, kde = False, fit_kws={"color": "k", "lw": 3},
            hist_kws={"linewidth": 4, 
                    "color": "g"})
ax.set_title("Variance: "+ str(np.std(Rs_norm)**2))
fig = ax.get_figure()
fig.savefig('adv_dist.png')
```

```python
ax = sns.distplot(Rs_norm * (1 - pis),fit=norm, kde = False, fit_kws={"color": "k", "lw": 3},
            hist_kws={"linewidth": 4, 
                    "color": "g"})
ax.set_title("Variance: "+ str(np.std(Rs_norm*(1 - pis))**2))
fig = ax.get_figure()
fig.savefig('adv_score_dist.png')
```
